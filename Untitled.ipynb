```python
from skimage.io import imread_collection, imshow
aseman_folder = 'data_batik/train/Aseman/*.jpg'
aseman = imread_collection(aseman_folder)

bledak_folder = 'data_batik/train/Aseman/*.jpg'
bledak = imread_collection(bledak_folder)

gunung_ringgit_folder = 'data_batik/train/GunungRinggit/*.jpg'
gunung_ringgit = imread_collection(gunung_ringgit_folder)

krecak_folder = 'data_batik/train/Krecak/*.jpg'
krecak = imread_collection(krecak_folder)

latohan_folder = 'data_batik/train/Latohan/*.jpg'
latohan = imread_collection(latohan_folder)

naga_folder = 'data_batik/train/Naga/*.jpg'
naga = imread_collection(naga_folder)

sekar_jagat_folder = 'data_batik/train/SekarJagat/*.jpg'
sekar_jagat = imread_collection(sekar_jagat_folder)
```

```python
def get_category_each_folder(dataset, label):
    temp_category=[]
    if label is 'aseman':
        temp_category = insert_numerical(dataset, temp_category, label)
        category_folder['aseman'].append(temp_category)
    elif label is 'bledak':
        temp_category = insert_numerical(dataset, temp_category, label)
        category_folder['bledak'].append(temp_category)
    elif label is 'gunung_ringgit':
        temp_category = insert_numerical(dataset, temp_category, label)
        category_folder['gunung_ringgit'].append(temp_category)
    elif label is 'krecak':
        temp_category = insert_numerical(dataset, temp_category, label)
        category_folder['krecak'].append(temp_category)
    elif label is 'latohan':
        temp_category = insert_numerical(dataset, temp_category, label)
        category_folder['latohan'].append(temp_category)
    elif label is 'naga':
        temp_category = insert_numerical(dataset, temp_category, label)
        category_folder['naga'].append(temp_category)
    else:
        temp_category = insert_numerical(dataset, temp_category, label)
        category_folder['sekar_jagat'].append(temp_category)    
    return category_folder

def initialize_results():
    from collections import defaultdict
    x_train = defaultdict(list)
    x_test = defaultdict(list)
    y_train = defaultdict(list)
    y_test = defaultdict(list)
    category_folder = defaultdict(list)
    return x_train, x_test, y_train, y_test, category_folder
    
def split_data(dataset, category_folder, category, x_test, x_train, y_test, y_train):
    from sklearn.model_selection import train_test_split
    for item in category:
        if item is 0:
            x_train['aseman'], x_test['aseman'], y_train['aseman'], y_test['aseman'] = train_test_split(aseman, category_folder['aseman'][0])
        elif item is 1:
            x_train['bledak'], x_test['bledak'], y_train['bledak'], y_test['bledak'] = train_test_split(bledak, category_folder['bledak'][0])
        elif item is 2:
            x_train['gunung_ringgit'], x_test['gunung_ringgit'], y_train['gunung_ringgit'], y_test['gunung_ringgit'] = train_test_split(gunung_ringgit, category_folder['gunung_ringgit'][0])
        elif item is 3:
            x_train['krecak'], x_test['krecak'], y_train['krecak'], y_test['krecak'] = train_test_split(krecak, category_folder['krecak'][0])
        elif item is 4:
            x_train['latohan'], x_test['latohan'], y_train['latohan'], y_test['latohan'] = train_test_split(latohan, category_folder['latohan'][0])
        elif item is 5:
            x_train['naga'], x_test['naga'], y_train['naga'], y_test['naga'] = train_test_split(naga, category_folder['naga'][0])
        else:
            x_train['sekar_jagat'], x_test['sekar_jagat'], y_train['sekar_jagat'], y_test['sekar_jagat'] = train_test_split(sekar_jagat, category_folder['sekar_jagat'][0])
    return x_train, x_test, y_train, y_test

def insert_numerical(dataset,category,label):
    for i in range(0,len(dataset)):
        if label is 'aseman':
            category.append(0)
        elif label is 'bledak':
            category.append(1)
        elif label is 'gunung_ringgit':
            category.append(2)
        elif label is 'krecak':
            category.append(3)
        elif label is 'latohan':
            category.append(4)
        elif label is 'naga':
            category.append(5)
        else:
            category.append(6)
    return category

def append_raw_data(dataset, raw_data, label):
    for i in range(0, len(dataset)-1):
        if label is 'aseman':
            raw_data.append(aseman[i])
        elif label is 'bledak':
            raw_data.append(bledak[i])
        elif label is 'gunung_ringgit':
            raw_data.append(gunung_ringgit[i])
        elif label is 'krecak':
            raw_data.append(krecak[i])
        elif label is 'latohan':
            raw_data.append(latohan[i])
        elif label is 'naga':
            raw_data.append(naga[i])
        else:
            raw_data.append(sekar_jagat[i])
    return raw_data
```
